```python
# Change the column name
df2 = df.rename(columns = {'Indicator': 'Indicator_ID'}, inplace = False)
df2.head(5)
```

```python
df.rename(columns = {'Indicator': 'Indicator_ID'}, inplace = True)
df.head(5)
```

```python
# Change names of multiple columns
df.rename(columns = {'PUBLISH STATES': 'Publication Status', 'WHO region':'WHO Region' }, inplace = True)
df.head(5)
```

```python
# Arrange values of a particular column in ascending order
df.sort_values(by = 'Year', inplace = True)
df.head(5)
```

```python
df3 = (df.sort_values(['Indicator_ID','Country', 'WHO Region'])).sort_index()
df3 = df3[['Indicator_ID', 'Country', 'Year', 'WHO Region', 'Publication Status']]
df3.head(3)
```
